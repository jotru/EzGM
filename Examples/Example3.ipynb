```python
from EzGM.Selection import ec8_part1
from time import time
from EzGM.Utility import RunTime
```

```python
# Acquire the run start time
startTime = time()
```

### 1) Initialize the ec8_part1 object for record selection

```python
spec = spec = ec8_part1(database='NGA_W2', outdir='Outputs')
```

## 2) Select the ground motions

```python
spec.select(ag=0.2,xi=0.05, I=1.0, Type='Type1',Soil='A', nGM=11, selection=1, Tp=1,
           Mw_lim=[6.5, 8], Vs30_lim=[200, 700], Rjb_lim=[0, 20], fault_lim=None, opt=0, 
           maxScale=2)

# selected records can be plotted at this stage
spec.plot(save=1, show=1)
```

### 4) If database == 'NGA_W2' you can first download the records via nga_download method from NGA-West2 Database [http://ngawest2.berkeley.edu/] and then use write method

### 5) If you have records already inside recs_f\database.zip\database or downloaded records for database = NGA_W2 case, write whatever you want, the object itself, selected and scaled time histories

```python
spec.write(obj=1, recs=0, recs_f='')
```

```python
# Calculate the total time passed
RunTime(startTime)
```
